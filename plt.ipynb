```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
plt.plot(x, x, label = 'linear')
plt.legend()
plt.show()
```

```python
data = {
    'Country' : ['Brazil', 'Russia', 'India', 'Japan'],
    'Population' : [200.4, 143.5, 450.12, 52.98]
}
import pandas as pd
df = pd.DataFrame(data)
plt.xlabel('Country')
plt.ylabel('Population')
plt.plot(df['Country'], df['Population'])
plt.legend()
plt.show()
```
