```python
df['Dep_Min'] = pd.to_datetime(df['Dep_Time']).dt.minute
```

```python
df['Dep_Hour'] = pd.to_datetime(df['Dep_Time']).dt.hour
```

```python
df['Arrival_Hour'] = pd.to_datetime(df['Arrival_Time']).dt.hour
```

```python
df['Arrival_Min'] = pd.to_datetime(df['Arrival_Time']).dt.minute
```

```python
df['Journey_Day'] = pd.to_datetime(df['Date_of_Journey']).dt.day
```

```python
df['Journey_Month'] = pd.to_datetime(df['Date_of_Journey']).dt.month
```

```python
df.drop(['Arrival_Time', 'Dep_Time', 'Date_of_Journey'], axis=1, inplace=True)
```

```python
df.info()
```

```python
df['Duration'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
import numpy as np

encoder = LabelEncoder()

df['Airline'] = encoder.fit_transform(df['Airline'])
airline_mapping = dict(zip(encoder.classes_, encoder.transform(encoder.classes_)))
print(airline_mapping)

cols = ['Source', 'Destination', 'Additional_Info', 'Route', 'Total_Stops', 'Duration']

for col in cols:
    df[col] = encoder.fit_transform(df[col])
```

```python
df.info()
```

```python
airlines_counts = df['Airline'].value_counts()
print(airlines_counts)
```

```python
import matplotlib.pyplot as plt


plt.pie(airlines_counts, labels=airlines_counts.index, autopct='%1.1f%%')
plt.title('Airline Distribution')
plt.show()

import seaborn as sns

plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True)
plt.show()

```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.model_selection import train_test_split
x = df.drop('Price', axis = 1)
y = df['Price']

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
X_train_1, X_test_1, y_train_1, y_test_1 = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
reg = LinearRegression()
reg.fit(X_train, y_train)
reg.score(X_test, y_test)
```

```python
reg.fit(X_train_1, y_train_1)
reg.score(X_test_1, y_test_1)
```

```python
rfor = RandomForestRegressor()
rfor.fit(X_train, y_train)
rfor.score(X_test, y_test)
```

```python
rfor.fit(X_train_1, y_train_1)
rfor.score(X_test_1, y_test_1)
```
